# 2. Revisión tabla: geolocations

```python
# visualizacion del dataframe
geolocations.head()
```

```python
# Visualización de la forma del dataframe.
geolocations.shape
```

```python
# visualización de información del dataframe.
geolocations.info()
```

```python
# visualización de valores nulos en el dataframe.
geolocations.isnull().sum()
```

```python
# visualización de valores duplicados en el dataframe.
geolocations.duplicated().sum()
```

```python
# eliminar duplicados exactos
geolocations = geolocations.drop_duplicates().copy()
geolocations.duplicated().sum()
```

```python
#  Eliminar duplicados exactos y crear nueva variable geolocations_clean
geolocations_clean = geolocations.drop_duplicates().copy()

#  Ver si siguen existiendo múltiples filas por código postal
geolocations_clean["geolocation_zip_code_prefix"].duplicated().sum()
```

```python
#  Agrupar por código postal y calcular la media de latitud y longitud, y tomar la primera ciudad y estado de los duplicados.
geolocations_zip = (
    geolocations_clean
    .groupby("geolocation_zip_code_prefix", as_index=False)
    .agg({
        "geolocation_lat": "mean",
        "geolocation_lng": "mean",
        "geolocation_city": "first",
        "geolocation_state": "first"
    })
)

geolocations_zip.shape
```

```python
# Ver si siguen existiendo múltiples filas por código postal
geolocations_zip["geolocation_zip_code_prefix"].duplicated().sum()
```

#### Limpieza y tratamiento de la tabla `geolocations`

La tabla `geolocations` contiene información geográfica asociada a prefijos de código postal. A diferencia de otras tablas del dataset, esta tabla no está vinculada directamente a un identificador individual de cliente, sino que puede relacionarse con la tabla `customers` mediante la coincidencia entre `customer_zip_code_prefix` y `geolocation_zip_code_prefix`.

En la revisión inicial no se detectaron valores nulos, pero sí un número elevado de duplicados exactos. Estos duplicados correspondían a filas que repetían la misma combinación de prefijo postal, latitud, longitud, ciudad y estado, por lo que fueron considerados registros redundantes y eliminados sin pérdida de información analítica.

Tras eliminar los duplicados exactos, se comprobó que muchos prefijos de código postal seguían apareciendo varias veces con coordenadas distintas. Esto indica que un mismo prefijo postal puede estar asociado a varias observaciones geográficas dentro de una misma zona. Para evitar que una futura unión con la tabla de clientes multiplicase artificialmente el número de registros, se construyó una versión agregada de la tabla con una sola fila por prefijo postal.

En esta tabla agregada, cada `geolocation_zip_code_prefix` queda representado por la latitud y longitud medias de sus observaciones, conservando además una ciudad y un estado de referencia. El resultado fue una tabla con 19.015 prefijos postales únicos y sin duplicados en la variable utilizada para la unión.

Es importante señalar que estas coordenadas no representan la ubicación exacta del domicilio del cliente, sino una localización aproximada asociada a su zona postal. Por tanto, su uso es adecuado para análisis territoriales agregados, visualización geográfica o estimación aproximada de distancias logísticas, pero no para una geolocalización individual precisa.

# Síntesis general de la limpieza realizada

A lo largo del notebook se realizó una revisión inicial de las principales tablas del dataset de e-commerce: `customers`, `geolocations`, `order_items`, `order_payments`, `order_reviews`, `order_dataset`, `products` y `sellers`.

En todas las tablas se aplicó un procedimiento común de exploración: revisión de las primeras filas, dimensiones del dataframe, tipos de datos, valores nulos, duplicados exactos y coherencia básica de las variables. Esta revisión permitió identificar qué tablas requerían transformaciones específicas y cuáles presentaban ya una estructura suficientemente limpia.

Las transformaciones aplicadas fueron principalmente de cuatro tipos. En primer lugar, se normalizaron variables textuales como ciudades y estados, eliminando espacios innecesarios y homogeneizando mayúsculas y minúsculas. En segundo lugar, se convirtieron variables temporales almacenadas como texto a formato `datetime`, como `shipping_limit_date`, `review_creation_date` y `review_answer_timestamp`, para permitir análisis temporales posteriores.

En tercer lugar, se trataron variables categóricas codificadas como numéricas, especialmente los prefijos de código postal, que fueron interpretados como identificadores geográficos y no como magnitudes cuantitativas. En cuarto lugar, se revisaron variables económicas y físicas para detectar valores negativos, valores iguales a cero o posibles incoherencias. En el caso de `products`, los pesos iguales a cero fueron convertidos a valores ausentes, al considerarse incompatibles con productos físicos.

La tabla `geolocations` requirió un tratamiento específico, ya que contenía un elevado número de duplicados exactos y múltiples coordenadas para un mismo prefijo postal. Para evitar problemas en futuras integraciones, se eliminaron duplicados exactos y se construyó una tabla agregada por código postal, usando la latitud y longitud medias como localización aproximada de cada zona postal.

En algunos casos se conservaron valores nulos por razones metodológicas. Por ejemplo, en `order_reviews`, los nulos en título o comentario de la reseña no se consideran errores, sino ausencia de texto por parte del cliente. Del mismo modo, en `products`, los nulos catalográficos se mantuvieron para no eliminar productos que podrían estar asociados a pedidos reales. En `order_dataset`, los nulos en fechas logísticas deben interpretarse con cautela, ya que pueden representar eventos no ocurridos o no registrados, como pedidos no aprobados, no enviados o no entregados.

En conjunto, la limpieza realizada prepara los datos para la siguiente fase del proyecto: la integración de tablas y la construcción de variables analíticas orientadas al rendimiento logístico, como tiempos de entrega, retrasos, costes de envío, características físicas del producto, ubicación aproximada de clientes y vendedores, y satisfacción del cliente.
