# Sistema de relaciones del modelo relacional

Este documento resume las relaciones principales entre las tablas del dataset y explica la lógica de claves primarias, claves foráneas y cardinalidades que se utilizará en el modelo relacional.

## 1. Regla general para definir claves foráneas

En una relación de tipo `1:N`, la clave foránea se coloca en la tabla que está en el lado `N`.

Es decir:

    tabla_padre 1 ─── N tabla_hija

La tabla hija contiene la clave foránea que apunta a la clave primaria de la tabla padre.

Ejemplo:

    customers_dataset 1 ─── N orders_dataset

Un cliente puede tener varios pedidos, pero cada pedido pertenece a un solo cliente. Por tanto, la clave foránea va en `orders_dataset`.

    orders_dataset.customer_id → customers_dataset.customer_id

## 2. Relación entre `customers_dataset` y `orders_dataset`

    customers_dataset 1 ─── N orders_dataset

Un cliente puede tener varios pedidos, mientras que cada pedido pertenece a un único cliente.

La tabla padre es `customers_dataset`.

La tabla hija es `orders_dataset`.

La clave primaria está en:

    customers_dataset.customer_id

La clave foránea está en:

    orders_dataset.customer_id

Relación SQL:

    FOREIGN KEY (customer_id)
    REFERENCES customers_dataset(customer_id)

## 3. Relación entre `orders_dataset` y `order_items_dataset`

    orders_dataset 1 ─── N order_items_dataset

La tabla padre es `orders_dataset`.

La tabla hija es `order_items_dataset`.

La clave primaria está en:

## 10. Relaciones muchos a muchos resueltas mediante `order_items_dataset`

Aunque muchas relaciones directas del modelo son `1:N`, existen relaciones conceptuales de tipo `N:M` que se resuelven mediante la tabla `order_items_dataset`.

### Relación entre pedidos y productos

Conceptualmente:

    orders_dataset N ─── M products_dataset

Un pedido puede contener muchos productos y un producto puede aparecer en muchos pedidos.

Esta relación muchos a muchos se resuelve mediante:

    orders_dataset 1 ─── N order_items_dataset N ─── 1 products_dataset

Por tanto, en la implementación relacional no se crea una relación directa muchos a muchos entre `orders_dataset` y `products_dataset`, sino que se utiliza `order_items_dataset` como tabla intermedia o tabla de detalle.

### Relación entre pedidos y vendedores

Conceptualmente:

    orders_dataset N ─── M sellers_dataset

Un pedido puede contener productos de varios vendedores y un vendedor puede aparecer en muchos pedidos.

Esta relación muchos a muchos se resuelve mediante:

    orders_dataset 1 ─── N order_items_dataset N ─── 1 sellers_dataset

Por tanto, en la implementación relacional no se crea una relación directa muchos a muchos entre `orders_dataset` y `sellers_dataset`, sino que se utiliza `order_items_dataset` como tabla intermedia o tabla de detalle.

## 13. Orden lógico de creación de tablas

El orden de creación de las tablas no depende de cuál sea más importante desde el punto de vista del análisis, sino de las dependencias entre ellas.

Primero deben crearse las tablas que no dependen de otras tablas mediante claves foráneas. Después se crean las tablas que apuntan a ellas.

Un posible orden lógico sería:

    1. geolocations_zip
    2. customers_dataset
    3. sellers_dataset
    4. products_dataset
    5. orders_dataset
    6. order_items_dataset
    7. order_payments_dataset
    8. order_reviews_dataset

La razón es la siguiente:

`geolocations_zip` debe existir antes que `customers_dataset` y `sellers_dataset`, porque ambas tablas apuntan a ella mediante el prefijo postal.

`customers_dataset` debe existir antes que `orders_dataset`, porque `orders_dataset.customer_id` apunta a `customers_dataset.customer_id`.

`orders_dataset`, `products_dataset` y `sellers_dataset` deben existir antes que `order_items_dataset`, porque `order_items_dataset` apunta a las tres mediante `order_id`, `product_id` y `seller_id`.

`orders_dataset` debe existir antes que `order_payments_dataset` y `order_reviews_dataset`, porque ambas tablas apuntan a ella mediante `order_id`.

## 14. Idea clave

La clave primaria identifica de forma única cada registro dentro de su propia tabla.

La clave foránea conecta una tabla dependiente con otra tabla principal.

En una relación `1:N`, la clave foránea se coloca siempre en la tabla del lado `N`.

Por tanto, el modelo no se organiza según qué tabla sea más importante, sino según las dependencias entre entidades y la cardinalidad de sus relaciones.

El modelo relacional permite conservar la estructura original del dataset, mantener la trazabilidad entre entidades y evitar duplicaciones artificiales. Posteriormente, sobre esta capa relacional podrá construirse una capa analítica dimensional orientada al cálculo de indicadores, visualización y predicción.
